```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
import SBM_generation as sbm
```

```python
K = 2
N = 500
avg_degree = 10
folder = './data/'
a = 0.1
```

```python
gm = sbm.SyntNetSBM(N=N,K=K,avg_degree=avg_degree,output_parameters=True,folder=folder,verbose=1,a=a)
A,G = gm.synthetic_network()
Ad = A.todense()
```

```python
infile = folder + 'N493K2_theta_syn_test_10.npz'
theta = np.load(infile)
u = theta['u']
w = theta['w']
M = np.einsum('ik,jq->ijkq', u, u)
M = np.einsum('ijkq,kq->ij', M, w)
```

```python
cmap = 'PuBuGn'
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.imshow(Ad, cmap = plt.get_cmap(cmap))
plt.subplot(1,2,2)
plt.imshow(M, cmap = plt.get_cmap(cmap))
plt.show()
```
